```python
# Dropping features with high count of null values
tr_feat = tr_feat.drop(['scheme_name','funder','installer','public_meeting','scheme_management',
                                'permit'], axis=1)

tr_feat.isnull().sum()
```

```python
# Create train validate test split
from sklearn.model_selection import train_test_split

X_train = tr_feat
y_train = tr_labels['status_group']

X_train, X_val, y_train, y_val = train_test_split(
    X_train, y_train, train_size=0.80, test_size=0.20, 
    stratify=y_train, random_state=56)

X_train.shape, X_val.shape, y_train.shape, y_val.shape
```

```python
# Load category encoders module
import category_encoders as ce
```

```python
# Apply scaling to increase accuracy of the model fit
# This requires another function from sklearn library
from sklearn.preprocessing import StandardScaler
```

```python
# Apply scaler method to train and validation sets
# Create feature list
categorical_features = ['quality_group', 'quantity_group','extraction_type','waterpoint_type','source','water_quality','management','payment','region']
numeric_features = X_train.select_dtypes('number').columns.drop(['id','num_private','district_code','construction_year']).tolist()
features = categorical_features + numeric_features

# Create train and validations subsets
X_train_subset = X_train[features]
X_val_subset = X_val[features]

# Use one hot encoding as above
encoder = ce.OneHotEncoder(use_cat_names=True)

# For training set we use fit_transform
X_train_encoded = encoder.fit_transform(X_train_subset)

# For validation set use transform
X_val_encoded = encoder.transform(X_val_subset)

# Apply scaling to scaler
scaler = StandardScaler()

# Apply scaler to train and validation sets
X_train_scaled = scaler.fit_transform(X_train_encoded)
X_val_scaled = scaler.transform(X_val_encoded)
```

```python
# Train logistic regression model using train data
from sklearn.linear_model import LogisticRegression
model = LogisticRegression(solver='lbfgs', multi_class='auto', max_iter=10000)
model.fit(X_train_scaled, y_train)
```

```python
# Calculate accuracy of model
print('Model accuracy using validation data is:', model.score(X_val_scaled, y_val))
```

```python
# Run model on test data set
X_test_subset = te_feat[features]
X_test_encoded = encoder.transform(X_test_subset)
X_test_scaled = scaler.transform(X_test_encoded)
```

```python
# Verify columns match between train and test
all(X_test_encoded.columns == X_train_encoded.columns)
```

```python
# Create submission file
sub = submission.copy()
sub['status_group'] = model.predict(X_test_scaled)
sub.to_csv('NDoshi_DS4_Sub.csv', index = False)
```
